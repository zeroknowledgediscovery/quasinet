```python
from quasinet.qnet import Qnet, save_qnet, load_qnet
import pandas as pd
```

```python
df=pd.read_csv('ptsdALPHnoIndex.txt', keep_default_na=False).sample(50)
```

```python
df
```

```python
feature_names=df.columns.values
```

```python
feature_names
```

```python
X=df.values.astype('U10')
```

```python
model = Qnet(n_jobs=5,feature_names=feature_names,alpha=.15)
```

```python
model.fit(X)
```

```python
save_qnet(model,'model',gz=True)
```
